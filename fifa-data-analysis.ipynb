```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Завантажуємо дані з папки data
df = pd.read_csv('data/top_prospects.csv')

# Налаштовуємо розмір графіка
plt.figure(figsize=(10, 6))

# Будуємо графік (тепер без зайвих дужок і з коректними параметрами)
sns.barplot(
    x='growth_needed', 
    y='short_name', 
    data=df, 
    hue='short_name', 
    palette='viridis', 
    legend=False
)

# Оформлення
plt.title('Топ-10 гравців за потенціалом росту', fontsize=15)
plt.xlabel('Одиниць росту (Potential - Overall)')
plt.ylabel('Гравець')
plt.grid(axis='x', linestyle='--', alpha=0.6) # додаємо легку сітку для зручності
plt.show()
```
